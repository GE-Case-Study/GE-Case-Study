## Creating a Pandas Profile for GE

```python
GE = pd.read_csv('ge_cleaned_data_corrected.csv')
```

```python
GE.describe()
```

```python
ge_profile = ProfileReport()
ge_profile.config.interactions.targets = ["rul"]
```

```python
# focuses only on interactions with RUL
ge_profile.df = GE
ge_profile.to_widgets()
```

# Summary  #

##### Overview 
Focusing on the 'Overview' aspect of the profile generated by pandas, we can see that when the variable 'rul' is made the independent variable, there were several variables that appear constant throughout the dataset and hold high correlation to other variables present. This part of the profile gave us a starting postiion when determining which variables are relevant to determining RUL and which insignificant variables could be dropped when modeling the data.

##### Interactions 
Moving on to how these variables correlate with each other, we noticed a few interesting relationships under 'Interactions'. With the calculation of RUL in mind, we focused on the variables that not only showed change but showed a normal distribution among the dataset. This brought our focus onto the variables 'distance', 'flight cycle', 't30', 't50', 'ps30', 'bpr', 'w32', 'nf', 'hpc_eff_mod'. Looking at each interaction, we can see that as RUL increased or decreased, there was a change in the dependent variable, making it a possible predictor of 'rul'.

##### Correlations 
Looking at the correlation heat map automatically generated by the profile, we can see that very few variables actually correlated with our target variable 'RUL'. Those with a dark blue hue to symbolized a corrolation level of 1 are only 'dataset', 'tra', 't2', 'p2', 'epr', 'farb', 'nf_dmd' and 'pcnfr_dmd'. However, these can be ruled out as significant due to these variables staying constant throughout the dataset, making them unimportant in our analysis of predicting 'rul'. Variables that do seem to have some leveage with a hint of a blue hue AND make sense pertaining to our analysis are 'operator', 'w31', 'w32', 'phi' and 'p30'. Each one doesn't appear to have a high correlation with 'rul' by themselves, but peaks the possibility of significance when grouped with other slightly correlated variables.

##### Missing Values 
The information provided under 'Missing Values' shows up which variables we either need to drop from our model, or need to have values created by using a form of imputation. The variables shown to have NULL values are 'distance', 'dest_latitude', 'dest_longitude', and 'rul'. With 'distance' being computed by using the variables 'dest_latitude' and 'dest_longitude', it makes sense that there would be missing data from these three columns as there is the chance an airplane did not leave and thus its destination airpline location is the same as its last departing airline location. Due to the significance of data provided by each column that contains missing information, we concluded to impute the missing data as we could assume its value and avoid a dataset with holes when predicting 'rul'.
